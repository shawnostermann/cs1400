```python
def beer_song(num):
    """ Recursively sing the beer bottle song
        num must be a positive integer 
    """
    print ()                                                    # print a blank line
    print (num, 'bottles of beer on the wall')
    print (num, 'bottles of beer')
    print ('If one of those bottles should happen to fall')
    next = num-1                                           # one less bottle
    if next > 0:                                           # keep singing
        if next>1:
            plural="bottles"
        else:
            plural="bottle"
        print (next, plural, 'of beer on the wall!')
        beer_song(next)                                       # recursive call to beer_song 
    else:                                                     # end the song
        print ('No more bottles of beer on the wall.') 

beer_song(20)
```
