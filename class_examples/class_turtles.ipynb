```python
def SHOW(c):
    clear_output(wait=True)
    display(c)
    time.sleep(1)
canvas = Canvas(size=(400, 400))
bob = Turtle(canvas, (200, 200), 0)

for i in range(4):   # this says to repeat the following code 4 times
    bob.forward(100)          # the code to be repeated is indented 
    bob.left(90)
    SHOW(canvas)
```

```python
def SHOW(c):
    clear_output(wait=True)
    display(c)
    time.sleep(1)
canvas = Canvas(size=(400, 400))
bob = Turtle(canvas, (200, 200), 0)

def square(t, length):      # now square works for any size square
    for i in range(4):
        t.forward(length)   #  note the use of the parameter length
        t.left(90)
        SHOW(canvas)

square(bob, 100)         # call square with bob and 100 as arguments
square(bob, 50)         # call square with bob and 100 as arguments
```

```python
def SHOW(c):
    clear_output(wait=True)
    display(c)
    time.sleep(0.25)
canvas = Canvas(size=(400, 400))
bob = Turtle(canvas, (200, 400), 0)

def polygon(t, n, length):   # n is the number of sides
    angle = 360 / n              # determine the correct angle to move
    for i in range(n):            # repeat for all n sides
        t.forward(length)
        t.left(angle)                 # use the new local variable angle
        SHOW(canvas)

polygon(bob, 4, 100)           # call polygon with bob, 4, and 100 
                                 # as arguments
polygon(bob, 7, 70)              # or let bob take a different walk!
polygon(t=bob, length=90, n=9)             # or let bob take a different walk!

```

```python
def SHOW(c):
    clear_output(wait=True)
    display(c)
    time.sleep(0.1)
canvas = Canvas(size=(400, 400))
bob = Turtle(canvas, (200, 400), 0)

def polygon(t, n, length):   # n is the number of sides
    angle = 360 / n              # determine the correct angle to move
    for i in range(n):            # repeat for all n sides
        t.forward(length)
        t.left(angle)                 # use the new local variable angle
        SHOW(canvas)
        
import math                            # so that we can use pi 
def circle(t, r):                         # t is the turtle and r is the radius of the circle
       circumference = 2 * math.pi * r    # compute the circumference, 2πr
       n = 50                             # use 50 sides
       length = circumference / n         # compute the length of each side
       polygon(t, n, length)		      # draw the (approximate) circle			
circle(bob, 150) 
```

```python
def polyline(t, n, length, angle): # pass in the angle as a fourth parameter
       for i in range(n):
             t.forward(length)
             t.left(angle)
             SHOW(canvas)
            
def polygon(t, n, length):
       angle = 360 / n
       polyline(t, n, length, angle)

def arc(t, r, angle):
       arc_length = 2 * math.pi * r * angle / 360
       n = int(arc_length / 3) + 1
       step_length = arc_length / n
       step_angle = angle / n
       polyline(t, n, step_length, step_angle)

def circle(t, r):              # get a turtle and a radius
      arc(t, r, 360)           # draw a 360 degree arc!

```
