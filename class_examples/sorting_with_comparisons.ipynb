```python
# Ostermann's solution to the "sort the 3 names" problem from class last week
# This is as good as it can get in terms of "computer work" ("optimal"), 
# because it requires no more than 3 comparisons to solve the problem
import random as rand

for LOOPER in range(100):
    people = ["noname", "noname", "noname"]
    # now make up 3 random names
    people[0] = int(rand.random()*100)
    people[1] = int(rand.random()*100)
    people[2] = int(rand.random()*100)
    print("the 2 people are: ", people)
    
    if people[0] < people[1]:
        # (0,1,2) or (2,0,1) or (0,2,1)
        if people[1] < people[2]:
            order=(0,1,2)
        else:
            # (2,0,1) or (0,2,1)
            if people[2] < people[0]:
                order=(2,0,1)
            else:                
                order=(0,2,1)
    # 1>0 from the first comparison, so options left are (1,2,0) (1,0,2) and (2,1,0)
    elif people[1] < people[2]:
        # possibilities are (1,2,0) or (1,0,2) 
        if people[2] < people[0]:
            order=(1,2,0)
        else:
            order=(1,0,2)
    else:
        order=(2,1,0)
    print("  Correct order is ", people[order[0]], people[order[1]], people[order[2]])
    if (people[order[0]] > people[order[1]]) or (people[order[0]] > people[order[2]]) or (people[order[1]] > people[order[2]]):
        print("OOPS!!!!  That's not right!")
    
```
